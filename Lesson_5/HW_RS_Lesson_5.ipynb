1. Ответьте на вопросы:  
В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?  
Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы.

**Ответ:**

Колаборативная фильтрация является состаной частью гибридной рекоммендатаельной системы.

**Hybride = Content-based + Colaborative**

 Данная модель производит подбор рекоммендованных товаров основываясь на степени их схожести с другими товарами (item-based), либо исходя из схожести предпочтений пользователей (user-based). 

Однако, помимо этого, существует большое колличество дополнительных данных, т.н. контентной информации, которые колаборативняая фильтрация не учитывает. 

Имеется ввиду, например: частота посещения магазина, средний чек, возраст посетителя, сезонность товара,  индивидуальные параметры(теги) и.т.д. Эта так называемая контентная информация позволяет дополнить информацию о покупателях и пользователе и товаре, улучшая т.о. качество рекоммендаций. 

Гибриндые рекоммендательные системы частично позволяют решить проблемму холодного старта для покупателей, и при вводе в оборот нового товара, Ну, например при составлении рекоомендаций для ассортимента товаров весна - лето в магазине одежды, решая проблемму засчет введения новых фитчей.

Какие основные отличия предложенной системы от тех подходов, которые мы разбирали на семинарах? Какие проблемы могут возникнуть при выводе такой модели в продакшен?

**Ответ:**

В первую очередь следует отметить то, что продуктом являются вакансии. Ососбенностью данного продукта является то, пользоваетель если он уже выбрал что-то то скорее всего уже не будет снова искать работу. И вакансии, тоже как появляются так и исчезают. Т.о. подбор рекоммендаций необходимо производить  на основе какой-то априроной информации о пользователе, полученной из его резюме. и информации и описания вакансии. 

Отличе данной рекоммендательной системы от **item-based** и **user-based**, в которых рекомендации производятся, сами знаете исходя из чего. Здесь рекоммендации будут формироваться исходя из того насколько пользователь по своим навыкам, отраженным в резюме похож на товар(вакансию). Если впереформулировать, ижется сходство между тегами, которые проставил пользоваетель в своем резюме и тегами отмеченными в писании вакансии.  

Плюс ко всему. Описанная в статье система, в большей степени поисковик с элементами рекомендательной системы нежели рекомендательная система в чистом виде. В которой, результаты поискового запроса отправляются в рекоомендательную систему, и та в свою очередь, отбирает из полученного списка наиболее подходящие для пользователя вакансии, основываясь на признаках из поискового запроса и даннных из резюме.
