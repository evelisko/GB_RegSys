**Ответ:**

Колаборативная фильтрация является состаной частью гибридной рекоммендатаельной системы.

**Hybride = Content-based + Colaborative**

 Данная модель производит подбор рекоммендованных товаров основываясь на степени их схожести с другими товарами (item-based), либо исходя из схожести предпочтений пользователей (user-based). 

Однако, помимо этого, существует большое колличество дополнительных данных, т.н. контентной информации, которые колаборативняая фильтрация не учитывает. 

Имеется ввиду, например: частота посещения магазина, средний чек, возраст посетителя, сезонность товара,  индивидуальные параметры(теги) и.т.д. Эта так называемая контентная информация позволяет дополнить информацию о покупателях и пользователе и товаре, улучшая т.о. качество рекоммендаций. 

Гибриндые рекоммендательные системы частично позволяют решить проблемму холодного старта для покупателей, и при вводе в оборот нового товара, Ну, например при составлении рекоомендаций для ассортимента товаров весна - лето в магазине одежды, решая проблемму засчет введения новых фитчей.
